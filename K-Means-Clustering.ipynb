# Importing Basic libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Importing Scikit-learn Packages

```python
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
```

# Creating a Randomly Generated Dataset

```python
np.random.seed(0)
```

```python
X, y = make_blobs(n_samples=7000, centers=[[5, 3], [-3, -1], [4, -3], [1, 2], [9, 5]], cluster_std=0.9)
```

# Plotting the Randomly Generated Dataset

```python
plt.scatter(X[:, 0], X[:, 1], marker='.')
```

# Setting up K-means

```python
k_means = KMeans(init = "k-means++", n_clusters = 5, n_init = 15)
```

# Fitting the K-means model

```python
k_means.fit(X)
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

# Plotting the Final Model and Clusters

```python
fig = plt.figure(figsize=(6, 4))
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))
ax = fig.add_subplot(1, 1, 1)

for k, col in zip(range(len(k_means.cluster_centers_)), colors):

    my_members = (k_means_labels == k)
    cluster_center = k_means_cluster_centers[k]
    
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())

plt.show()
```
